### Assignment :- K-means clustering using make_blob

#### Generate clusters using make_blobs

```python
from sklearn.datasets import make_blobs

data = make_blobs(n_samples=1000, centers=5)
```

```python
X = data[0]
X
```

```python
y = data[1]
y
```

#### Plot the generated data

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
plt.scatter(X[ : , 0 ], X[ : , 1 ], c=y)
plt.show()
```

#### Scaling the data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

#### Modeling using K-Means

```python
from sklearn.cluster import KMeans
from sklearn.metrics  import accuracy_score

k_means = KMeans(n_clusters=5, max_iter=50, n_init=5, random_state=40) #Applying best parameters according to GridSearchCV
k_means.fit(X, y)
prediction = k_means.predict(X)

# accuracy_score(y, prediction)
plt.figure(figsize=(10,8))
plt.scatter(X[:, 0], X[:, 1], c=prediction)
```

#### Using GridSearchCV to get best_parameters

```python
from sklearn.model_selection import GridSearchCV

params = { 'n_clusters' : range(1, 6), 'n_init' : range(5, 50, 5), 'max_iter' : range(50, 300, 50) }
gscv = GridSearchCV(k_means, params, cv=5)
gscv.fit(X, y)

gscv.best_params_
```
