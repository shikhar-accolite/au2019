### Using SVC model

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

train_X, test_X, train_y, test_y = train_test_split( train_df.drop("price_range", axis=1), train_df["price_range"], test_size=0.30, random_state=42 )

svc = SVC(kernel="linear", C=1.0)
svc.fit( train_X, train_y )
train_y_predict = svc.predict( train_X )
print( accuracy_score(train_y_predict, train_y) )

test_y_predict = svc.predict( test_X )
print( accuracy_score(test_y_predict, test_y) )
```

### SVC model with scaling

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

train_X, test_X, train_y, test_y = train_test_split( train_df.drop("price_range", axis=1), train_df["price_range"], test_size=0.30, random_state=42 )

scaler = StandardScaler()
train_X = scaler.fit_transform(train_X)
test_X = scaler.fit_transform(test_X)

svc = SVC(kernel="linear", C=10.0)
svc.fit( train_X, train_y )
train_y_predict = svc.predict( train_X )
print( accuracy_score(train_y_predict, train_y) )

test_y_predict = svc.predict( test_X )
print( accuracy_score(test_y_predict, test_y) )
```

### Using RandomForest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler, MinMaxScaler

train_X, test_X, train_y, test_y = train_test_split( train_df.drop("price_range", axis=1), train_df["price_range"], test_size=0.30, random_state=42 )

# scaler = StandardScaler()
# train_X = scaler.fit_transform(train_X)
# test_X = scaler.fit_transform(test_X)

rfc = RandomForestClassifier()
rfc.fit( train_X, train_y )
train_y_predict = rfc.predict( train_X )
print( accuracy_score(train_y_predict, train_y) )

test_y_predict = svc.predict( test_X )
print( accuracy_score(test_y_predict, test_y) )
```
